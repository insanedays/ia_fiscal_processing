```python
import pandas as pd
import unicodedata
```

```python
header = pd.read_csv('../data/raw/202401_NFs_Cabecalho.csv')
itens = pd.read_csv('../data/raw/202401_NFs_itens.csv') 
```

```python
header_columns = header.columns
header_columns
```

```python
itens_columns = itens.columns
itens_columns
```

```python
# understanding if there are columns in itens that are not in header
drop_columns = header_columns.intersection(itens_columns).drop('CHAVE DE ACESSO') 
itens_columns_selected = itens_columns.drop(drop_columns)

itens_copy = itens.copy()

itens_silver = itens_copy[itens_columns_selected]
itens_silver.head(2)
```

```python
def normalize_col(col_name):
    #  remove accent marks, lowercase, and replace special characters
    col_name = unicodedata.normalize('NFKD', col_name).encode('ASCII', 'ignore').decode('utf-8')
    col_name = col_name.lower()
    col_name_new = col_name.replace(' ', '_') \
                       .replace('/', '_') \
                       .replace('(', '') \
                       .replace(')', '') \
                       .strip()
    
    print(f'{col_name_new} <- {col_name}')
    return col_name_new
```

```python
print('Mudança no campos da itens_silver:')
itens_silver.columns = [normalize_col(col) for col in itens_silver.columns]
header_silver = header.copy()
print('\nMudança no campos da header_silver:')
header_silver.columns = [normalize_col(col) for col in header_silver.columns]
```
